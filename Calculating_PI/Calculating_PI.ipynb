```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math
```

```python
m = [20,100,1000]
  
circle = 0
square = 0

for n in m:
  figure, ax1 = plt.subplots()
  for i in range(n):

      rand_x= random.uniform(0, 1)
      rand_y= random.uniform(0, 1)
      plt.plot(rand_x,rand_y,'.',color="orange")

      
      origin_dist= rand_x**2 + rand_y**2
    
      if origin_dist<= 1:
          circle += 1
    
      square += 1
    
      pi = 4* circle / square 
    
  print("Estimation of Pi for n =", n)
  print("is", pi) 

  x = np.linspace(0,1,100)
  y_list = []
  for i in x:
    y_list.append(math.sqrt(1 - i**2))
  y = np.array(y_list)
  ax1.set_aspect( 1 )
  plt.plot(x,y, color='b')
```
